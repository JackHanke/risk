````sage
# define exact RISK 
from numpy import zeros, set_printoptions

def t_1(a,b):
    sum=0
    if (a+b)%2 == 1 or a<0 or b<0: return 0
    for j in range(((a+b)/2)+1): sum += binomial(((a+b)/2),((a-b)/2)+(2*j))*binomial(((a-b)/2)+(2*j), j) * ((2611/(85*sqrt(910)))**(b-(2*j)))
    return sum * ((2275/7776)**(a/2)) * ((2890/7776)**(b/2)) / 7776

def t_2(a,b): 
    sum = 0
    for i in range(a+1): 
        sum += t_1(i,b)
    return sum

def t_3(a,b): 
    sum = 0
    for j in range(b+1): 
        sum += t_1(a,j)*((55/216)**(b-j))
    return sum

def t_4(a,b): 
    sum = 0
    for i_2 in range(a+1): 
        for i_1 in range(i_2+1):
            sum += t_1(i_1,b)*((49/144)**(i_2-i_1))
    return sum

def t_5(a,b): 
    sum = 0
    for i_2 in range(b+1):
        for i_1 in range(i_2+1):
            sum += t_1(a,i_1)*(((55/216)**(b-i_2))*(((sqrt(295/1296))**(i_2-i_1)) + ((-sqrt(295/1296))**(i_2-i_1))))
    return sum/2

def t(a,b): 
    val = -7776*t_1(a-2,b) - 7776*t_1(a-3,b) - 3240*t_1(a-2,b-1) - 3254*t_1(a-3,b-1) + \
        7776*t_2(a-2,b) - 2611*t_2(a-3,b-1) - 2275*t_2(a-4,b) + \
        7776*t_3(a-2,b) + 1260*t_3(a-2,b-1) - 2611*t_3(a-3,b-1) - 2890*t_3(a-2,b-2) -\
        (91385/216)*t_3(a-3,b-2) - (50575/108)*t_3(a-2,b-3) + \
        3240 * t_4(a-2,b-1) + (3045/2)*t_4(a-3,b-1) - (6965/12)*t_4(a-4,b-1) - \
        (2309125/5184) * t_4(a-5,b-1) - (557375/5184) * t_4(a-6,b-1) + \
        7776*t_5(a-3,b) + 3885*t_5(a-3,b-1) - (240005/72)*t_5(a-3,b-2) -\
        (1871275/1296)*t_5(a-3,b-3) + (46131625/279936)*t_5(a-3,b-4)
    return val

set_printoptions(suppress=True, precision=4, linewidth=300)
the_table = zeros((dim,dim))
for i in range(2,dim):
    for j in range(dim):
        the_table[i][j] = t(i,j)
print(the_table)
````

````sage
from random import random

# simulates RISK combat for a troops attacks b troops. Optional threshold to stop the attack if attacking force dips below a certain level
def combat_sim(a,b,threshold=0):
    x = random()
    if a==1 or b==0 or a<=threshold: return f"The combat ended with {a} attackers and {b} defenders."
    elif a==2 and b==1: 
        if x < (15/36): return combat_sim(a,b-1)
        else: return combat_sim(a-1,b)
    elif a==2 and b>=2: 
        if x < (53/216): return combat_sim(a,b-1)
        else: return combat_sim(a-1,b)
    elif a==3 and b==1: 
        if x < (125/216): return combat_sim(a,b-1)
        else: return combat_sim(a-1,b)
    elif a==3 and b>=2: 
        if x < (295/1296): return combat_sim(a,b-2)
        elif x < (715/1296): return combat_sim(a-1,b-1)
        else: return combat_sim(a-2,b)
    elif a>=4 and b==1: 
        if x < (855/1296): return combat_sim(a-1,b)
        else: return combat_sim(a,b-1)
    elif a>=4 and b>=2: 
        if x < (2890/7776): return combat_sim(a,b-2)
        elif x < (5501/7776): return combat_sim(a-1,b-1)
        else: return combat_sim(a-2,b)

print(combat_sim(30,30, threshold=10))
````
